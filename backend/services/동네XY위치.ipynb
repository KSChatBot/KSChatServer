```python
def get_weather_location_XY(sido, gungu, dong):
    print("Weather Location : {} {} {}".format(sido, gungu, dong))

    excel_file = '기상청18_동네예보 조회서비스_오픈API활용가이드_격자_위경도(20210106).xlsx'
    # excel_file_url = 'http://localhost:8888/notebooks/services/기상청18_동네예보 조회서비스_오픈API활용가이드_격자_위경도(20210106).xlsx'
    df_sheet = pd.read_excel(excel_file, sheet_name='최종업데이트파일_20210106')

    df_result = df_sheet[(df_sheet['1단계'].str.contains(sido, na=False)) & (df_sheet['2단계'].str.contains(gungu, na=False)) & (df_sheet['3단계'] == dong)]
    rows = df_result.shape[0]

    if rows > 0:
        X_value = df_result.loc[df_result.index[0],'격자 X']
        Y_value = df_result.loc[df_result.index[0],'격자 Y']
    else:
        X_value = -1
        Y_value = -1

    print("Weather Location X:{} Y:{}".format(X_value, Y_value))

    return X_value, Y_value
```

```python
sido = "서울"
gungu = "종로"
dong = "사직동"
X, Y = get_weather_location_XY(sido, gungu, dong)
print(X, Y)
```
